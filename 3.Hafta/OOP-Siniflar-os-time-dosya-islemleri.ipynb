**Gizli Üyeler**  
Python’da şimdiye kadar gördüğümüz ve yukarıda andığımız aleni üyelerin dışında, bir de gizli üyeler bulunur. Aleni üyelerin aksine gizli üyeler dışarıya açık değildir. Gizli üyelere, normal yöntemleri kullanarak sınıf dışından erişemeyiz.Burada __gizli adlı bir gizli sınıf niteliği tanımladık. Bu değişkenin yalnızca baş tarafında iki adet alt çizgi olduğuna, ancak uç tarafında alt çizgi bulunmadığına dikkat edin. İşte Python’da baş tarafında yukarıdaki gibi iki adet alt çizgi olan, ancak uç tarafında alt çizgi bulunmayan (veya yalnızca tek bir alt çizgi bulunan) bütün öğeler birer gizli üyedir. Dışarıya kapalı olan bu gizli üyelere, normal yöntemleri kullanarak sınıf dışından erişemezsiniz.Gizli üyelere sınıf içerisinden ulaşabilirsiniz.

Python'da yapıcı (constructor) ve yıkıcı (destructor) metodlar, sırasıyla `__init__` ve `__del__` adlarıyla tanımlanır. İşte yapıcı ve yıkıcı metodlara sahip basit bir sınıf örneği:

```python
class Araba:
    def __init__(self, marka, model, renk):
        self.marka = marka
        self.model = model
        self.renk = renk
        print(f"{self.renk} renkli {self.marka} {self.model} arabası oluşturuldu.")

    def __del__(self):
        print(f"{self.renk} renkli {self.marka} {self.model} arabası yok edildi.")

# Yapıcı metot aracılığıyla Araba sınıfından bir nesne oluşturulur
araba1 = Araba("Renault", "Clio", "Beyaz")

# Yıkıcı metot nesne yok edilirken otomatik olarak çağrılır
del araba1
```

Yukarıdaki örnekte, `__init__` metodu yapıcı metot olarak kullanılmıştır. Bu metot, bir nesne oluşturulduğunda otomatik olarak çağrılır ve nesnenin başlangıç durumunu ayarlamak için kullanılır. `__del__` metodu ise yıkıcı metot olarak kullanılmıştır. Bu metot, bir nesne yok edildiğinde (örneğin, `del` anahtar kelimesiyle veya nesne kullanım dışına çıktığında) otomatik olarak çağrılır ve nesnenin bellekten silinmesiyle ilgili işlemleri gerçekleştirmek için kullanılır.

**Soru: Banka Hesap Sınıfı**

Aşağıdaki sınıf tanımını kullanarak bir banka hesabı uygulaması geliştirin:

```python
class BankaHesabi:
    def __init__(self, hesap_sahibi, bakiye=0):
        self.hesap_sahibi = hesap_sahibi
        self.bakiye = bakiye

    def para_yatir(self, miktar):
        if miktar > 0:
            self.bakiye += miktar
            print(f"{self.hesap_sahibi} adlı hesaba {miktar} TL yatırıldı. Yeni bakiye: {self.bakiye} TL")
        else:
            print("Geçersiz miktar. Lütfen pozitif bir değer girin.")

    def para_cevir(self, miktar, diger_hesap):
        if miktar > 0 and self.bakiye >= miktar:
            self.bakiye -= miktar
            diger_hesap.bakiye += miktar
            print(f"{self.hesap_sahibi} adlı hesaptan {miktar} TL çekildi. Yeni bakiye: {self.bakiye} TL")
            print(f"{diger_hesap.hesap_sahibi} adlı hesaba {miktar} TL aktarıldı.")
        else:
            print("Geçersiz miktar veya yetersiz bakiye.")

# Test Kodu
ahmet_hesap = BankaHesabi("Ahmet Yılmaz", 1000)
mehmet_hesap = BankaHesabi("Mehmet Kaya", 500)

ahmet_hesap.para_yatir(200)
mehmet_hesap.para_cevir(300, ahmet_hesap)
```

Yukarıdaki sınıfı ve test kodunu kullanarak şunları yapın:

1. Her iki hesabı da oluşturun (birini "Ahmet Yılmaz" için, diğerini "Mehmet Kaya" için).

2. "Ahmet Yılmaz" adlı hesaba 200 TL yatırın.

3. "Mehmet Kaya" adlı hesaptan "Ahmet Yılmaz" adlı hesaba 300 TL aktarın.

4. Her adımda bakiyeleri doğru şekilde güncellenmiş şekilde çıktıları kontrol edin.

5. Geçersiz işlemler yaparak (örneğin, negatif miktarla para yatırma veya yetersiz bakiyeyle para çekme) sınıfın doğru şekilde hata mesajı verdiğini kontrol edin.

Bu test kodu ve sınıfı kullanarak, BankaHesabi sınıfınızın işlevselliğini doğru bir şekilde test edebilirsiniz. Sorularınız varsa çekinmeden sorabilirsiniz.

### Orneği yapınız.

1. **Basit Bir Modül Oluşturma:**
   
   Öncelikle, basit bir matematik modülü oluşturalım. Bu modül, basit matematik işlemleri gerçekleştirecek fonksiyonlar içerecek.

   ```python
   # matematik.py adında bir dosya oluşturun
   # Bu dosya içinde basit matematik işlemleri için fonksiyonlar tanımlayın

   def toplama(a, b):
       return a + b

   def cikarma(a, b):
       return a - b

   def carpma(a, b):
       return a * b

   def bolme(a, b):
       if b != 0:
           return a / b
       else:
           return "Sıfıra bölme hatası!"
   ```

2. **Modülü Kullanma:**

   Şimdi oluşturduğumuz `matematik.py` modülünü kullanalım. Bunun için başka bir Python dosyası oluşturarak bu modülü içe aktaracağız ve içindeki fonksiyonları kullanacağız.

   ```python
   # kullan.py adında bir dosya oluşturun
   # Bu dosyada matematik modülünü içe aktararak fonksiyonlarını kullanın

   import matematik  # matematik modülünü içe aktar

   # Matematik fonksiyonlarını kullanarak işlemler yap
   print("Toplama: ", matematik.toplama(5, 3))
   print("Çıkarma: ", matematik.cikarma(10, 4))
   print("Çarpma: ", matematik.carpma(7, 2))
   print("Bölme: ", matematik.bolme(15, 3))
   ```

3. **Modülün Çalıştırılması:**

   `kullan.py` dosyasını çalıştırdığınızda, matematik modülünün içindeki fonksiyonları kullanarak hesaplamalar yapacak ve sonuçları ekrana yazdıracaktır.

   ```bash
   $ python kullan.py
   Toplama:  8
   Çıkarma:  6
   Çarpma:  14
   Bölme:  5.0
   ```

Bu örneklerde gördüğünüz gibi, bir modül oluşturmak için basit bir dosya oluşturuyoruz ve içine istediğimiz fonksiyonları tanımlıyoruz. Daha sonra bu modülü kullanmak için başka bir Python dosyası oluşturup içe aktarıyoruz (`import` ile). Bu şekilde modül içinde tanımladığımız fonksiyonları kullanarak işlemler yapabiliyoruz. Bu modüler yapı, büyük projelerde kodu düzenli tutmak ve yeniden kullanılabilirlik sağlamak için oldukça kullanışlıdır.

Python'un `os` modülü, işletim sistemiyle etkileşimde bulunmak için bir dizi işlev sağlar. İşte `os` modülünün önemli fonksiyonlarından bazıları:

1. **Dosya ve Dizin İşlemleri:**
   - `os.listdir(path='.')`: Belirtilen dizindeki dosya ve klasörleri listeler.
   - `os.mkdir(path, mode=0o777, *, dir_fd=None)`: Belirtilen dizinde yeni bir klasör oluşturur.
   - `os.makedirs(name, mode=0o777, exist_ok=False)`: Yolu içeren tüm klasörleri oluşturur.
   - `os.rename(src, dst, *, src_dir_fd=None, dst_dir_fd=None)`: Bir dosyanın adını değiştirir veya taşır.
   - `os.remove(path, *, dir_fd=None)`: Belirtilen dosyayı siler.
   - `os.rmdir(path, *, dir_fd=None)`: Belirtilen klasörü siler.
   - `os.removedirs(path)`: Yolu içeren tüm klasörleri siler.

2. **Yol İşlemleri:**
   - `os.getcwd()`: Çalışma dizinini döndürür.
   - `os.chdir(path)`: Çalışma dizinini değiştirir.
   - `os.path.exists(path)`: Belirtilen yolun var olup olmadığını kontrol eder.
   - `os.path.isdir(path)`: Belirtilen yolun bir klasör olup olmadığını kontrol eder.
   - `os.path.isfile(path)`: Belirtilen yolun bir dosya olup olmadığını kontrol eder.
   - `os.path.join(path1, path2, ...)`: Yolları birleştirir ve doğru bir yol oluşturur.

3. **Çevresel Değişkenler ve Sistem Bilgisi:**
   - `os.getenv(key, default=None)`: Çevresel bir değişkenin değerini döndürür.
   - `os.environ`: Çevresel değişkenlerin sözlük temsilini döndürür.
   - `os.system(command)`: Komut istemini çalıştırır (daha güvenli alternatifler tercih edilmelidir).
   - `os.cpu_count()`: Sistemdeki CPU çekirdek sayısını döndürür.
   - `os.uname()`: Sistem bilgilerini döndürür (yalnızca Unix sistemlerinde).

Bu liste, `os` modülünün önemli ve sık kullanılan fonksiyonlarından bazılarını içermektedir. Bu fonksiyonlar, işletim sistemiyle etkileşime geçmek, dosya ve dizin işlemleri yapmak, çevresel değişkenleri kontrol etmek ve sistem bilgilerini almak gibi işlevler sunar.

### Dosya İşlemleri
Python'da dosya işlemleri, dosya oluşturma, yazma, okuma ve silme gibi işlemleri gerçekleştirmek için kullanılır. Dosya işlemleri için `open()` fonksiyonu kullanılır ve bu fonksiyonun farklı modları vardır. İşte Python'da dosya işlemlerinin özellikleri:

1. **Dosya Oluşturma ve Açma:**
   
   Dosya işlemleriyle ilgili işlemlere başlamadan önce, bir dosyayı açmamız gereklidir. Bunun için `open()` fonksiyonu kullanılır.

   ```python
   dosya = open("dosya_adı.txt", "mod")
   ```

   Burada "dosya_adı.txt", oluşturulacak veya açılacak dosyanın adını temsil eder. "mod" ise dosyanın hangi modda açılacağını belirtir. Örnek modlar şunlardır:
   - "r": Okuma modu (varsayılan).
   - "w": Yazma modu. Dosyayı oluşturur veya varsa içeriğini siler.
   - "a": Ekleme modu. Dosyanın sonuna veri ekler.
   - "x": Oluşturma modu. Dosyayı oluşturur, ancak varsa hata verir.
   - "b": Binary modu. Binary dosyaları işlemek için kullanılır.

2. **Dosyaya Yazma:**

   Dosyaya veri yazmak için `write()` metodu kullanılır.

   ```python
   dosya.write("Merhaba, dosya!")
   ```

3. **Dosyadan Okuma:**

   Dosyadan veri okumak için `read()` veya `readline()` metodu kullanılır.

   ```python
   icerik = dosya.read()  # Tüm içeriği okur
   satir = dosya.readline()  # Bir satır okur
   ```

4. **Dosyayı Kapatma:**

   Dosya işlemlerinden sonra dosyanın kapatılması önemlidir. Bu, bellek kaynaklarının serbest bırakılmasını sağlar.

   ```python
   dosya.close()
   ```

5. **Dosya Silme:**

   Bir dosyayı silmek için `os` modülündeki `remove()` fonksiyonu kullanılır.

   ```python
   import os
   os.remove("dosya_adı.txt")
   ```

6. **Dosya İçin Otomatik Kapatma (with):**

   Dosyayı açtıktan sonra işlemler bittiğinde otomatik olarak kapatmak için `with` ifadesi kullanılır.

   ```python
   with open("dosya_adı.txt", "r") as dosya:
       icerik = dosya.read()
   ```

Python paket yöneticisi, Python dilinde kullanılan harici kütüphanelerin, modüllerin ve bağımlılıkların yönetimini sağlayan bir araçtır. 

Pip, terminal veya komut istemcisinden kullanılır. Aşağıda pip ile sık kullanılan bazı komutlar ve bunların nasıl kullanılacağına dair örnekler verilmiştir:

1. **Paket Yükleme (`install`)**:
   ```bash
   pip install paket_adi
   ```
   Örnek: `pip install requests` komutu, "requests" adlı paketi yükler.

2. **Paket Güncelleme (`install --upgrade`)**:
   ```bash
   pip install --upgrade paket_adi
   ```
   Örnek: `pip install --upgrade requests` komutu, "requests" paketini günceller.

3. **Paket Kaldırma (`uninstall`)**:
   ```bash
   pip uninstall paket_adi
   ```
   Örnek: `pip uninstall requests` komutu, "requests" paketini kaldırır.

4. **Kurulu Paketleri Listeleme (`list`)**:
   ```bash
   pip list
   ```
   Bu komut, yüklü olan paketleri listeler.

5. **Paket Bilgisi (`show`)**:
   ```bash
   pip show paket_adi
   ```
   Örnek: `pip show requests` komutu, "requests" paketi hakkında bilgi verir.

6. **Paket Dosyası İndirme (`download`)**:
   ```bash
   pip download paket_adi
   ```
   Örnek: `pip download requests` komutu, "requests" paketinin dosyalarını indirir.

7. **Belirli Bir Sürüm Yükleme**:
   ```bash
   pip install paket_adi==surum_numarasi
   ```
   Örnek: `pip install requests==2.25.1` komutu, "requests" paketinin belirli bir sürümünü yükler.

8. **Gereksinimleri Dosyasından Paketleri Yükleme**:
   ```bash
   pip install -r gereksinimler.txt
   ```
   Örnek: `pip install -r requirements.txt` komutu, "requirements.txt" dosyasındaki gereksinimleri yükler.

Pip'in yanı sıra, Python için başka paket yöneticileri de mevcuttur. Örneğin, `conda` Anaconda dağıtımıyla birlikte gelir ve Python paketlerini yönetmek için kullanılır. Ancak, genel olarak Python topluluğu tarafından en çok tercih edilen ve kullanılan paket yöneticisi `pip`'tir.
