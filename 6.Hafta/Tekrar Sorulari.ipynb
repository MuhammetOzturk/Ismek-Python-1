**Soru:**

Bir müzik kütüphanesi oluşturmak istiyorsunuz. Bu kütüphanede şarkıların bilgilerini saklamak için bir `Songs` modulu altında `Song` sınıfı oluşturun. Her şarkı için aşağıdaki özellikleri içeren bir sınıf tanımı yapılmalıdır:
`from Songs.Song import Song` şeklinde çağrılacak.

1. Şarkı adı (`title`)
2. Sanatçı adı (`artist`)
3. Albüm adı (`album`)
4. Şarkı süresi (`duration`) (`random modulu kullanınız.`)

Ayrıca, bu müzik kütüphanesinde şarkıları depolamak ve yönetmek için bir `MusicLibraries` modulu altında `MusicLibrary` sınıfı oluşturun. Bu sınıf, şu özelliklere sahip olmalıdır:

1. Müzik kütüphanesindeki tüm şarkıları depolamak için bir liste (`songs`) tutmalıdır.
2. Yeni bir şarkı eklemek için bir `add_song` metodu olmalıdır.
3. Müzik kütüphanesindeki tüm şarkıları listelemek için bir `list_songs` metodu olmalıdır.
4. Belirli bir sanatçının şarkılarını listelemek için bir `list_songs_by_artist` metodu olmalıdır.
5. Rastgele şarkı seçme metodu `get_random_song` bulunmalıdır . Seçilen şarkıyı `play_song` ile çalmalıdır.
6. Çalan şarkıyı gösteren `play_song` metodu ile o an çalan şarkı ekrana yazdırılmalıdır.
7. Şarkıların depolandığı yolun saklandığı `library_path` genel niteliği tanımlanmalıdır.

Bu işlevleri gerçekleştirecek şekilde `Song` ve `MusicLibrary` sınıflarını oluşturun. Ardından, bu sınıfları kullanarak birkaç örnek şarkı ekleyin ve müzik kütüphanesini kullanarak şarkıları listeleme ve belirli bir sanatçının şarkılarını listeleme işlemlerini yapın.

Kullanılacak modüller:  Şarkı süresi için ve ratsgele seçim için `random` modulu kullanılmalıdır .
